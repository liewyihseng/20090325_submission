```python
import pandas as pd
from scipy.stats import pearsonr
from numpy.random import randn
import matplotlib.pyplot as plot

pd.set_option("display.max_rows", None, "display.max_columns", None)

num_feature = 6
num_target = 3

dataset = pd.read_csv("../Dataset/Choudhury_Dataset.csv")
features = dataset.iloc[:, 0: num_feature]
targets = dataset.iloc[:, num_feature: num_feature + num_target]
```

```python
# Calculating the Pearson's Correlation between features
x_1 = features.iloc[:, 0].to_numpy()
x_2 = features.iloc[:, 1].to_numpy()
x_3 = features.iloc[:, 2].to_numpy()
x_4 = features.iloc[:, 3].to_numpy()
x_5 = features.iloc[:, 4].to_numpy()
x_6 = features.iloc[:, 5].to_numpy()
```

```python
# Scatter Plots
corr, _ = pearsonr(x_1, x_2)
print("Pearsons correlation of x_1 and x_2: %.3f" % corr)
plot.scatter(x_1, x_2)
plot.title("x_1 and x_2")
plot.xlabel('x_1')
plot.ylabel('x_2')
plot.show()

corr, _ = pearsonr(x_1, x_3)
print("Pearsons correlation of x_1 and x_3: %.3f" % corr)
plot.scatter(x_1, x_3)
plot.title("x_1 and x_3")
plot.xlabel('x_1')
plot.ylabel('x_3')
plot.show()

corr, _ = pearsonr(x_1, x_4)
print("Pearsons correlation of x_1 and x_4: %.3f" % corr)
plot.scatter(x_1, x_4)
plot.title("x_1 and x_4")
plot.xlabel('x_1')
plot.ylabel('x_4')
plot.show()

corr, _ = pearsonr(x_1, x_5)
print("Pearsons correlation of x_1 and x_5: %.3f" % corr)
plot.scatter(x_1, x_5)
plot.title("x_1 and x_5")
plot.xlabel('x_1')
plot.ylabel('x_5')
plot.show()

corr, _ = pearsonr(x_1, x_6)
print("Pearsons correlation of x_1 and x_6: %.3f" % corr)
plot.scatter(x_1, x_6)
plot.title("x_1 and x_6")
plot.xlabel('x_1')
plot.ylabel('x_6')
plot.show()

corr, _ = pearsonr(x_2, x_3)
print("Pearsons correlation of x_2 and x_3: %.3f" % corr)
plot.scatter(x_2, x_3)
plot.title("x_2 and x_3")
plot.xlabel('x_2')
plot.ylabel('x_3')
plot.show()

corr, _ = pearsonr(x_2, x_4)
print("Pearsons correlation of x_2 and x_4: %.3f" % corr)
plot.scatter(x_2, x_4)
plot.title("x_2 and x_4")
plot.xlabel('x_2')
plot.ylabel('x_4')
plot.show()

corr, _ = pearsonr(x_2, x_5)
print("Pearsons correlation of x_2 and x_5: %.3f" % corr)
plot.scatter(x_2, x_5)
plot.title("x_2 and x_5")
plot.xlabel('x_2')
plot.ylabel('x_5')
plot.show()

corr, _ = pearsonr(x_2, x_6)
print("Pearsons correlation of x_2 and x_6: %.3f" % corr)
plot.scatter(x_2, x_6)
plot.title("x_2 and x_6")
plot.xlabel('x_2')
plot.ylabel('x_6')
plot.show()

corr, _ = pearsonr(x_3, x_4)
print("Pearsons correlation of x_3 and x_4: %.3f" % corr)
plot.scatter(x_3, x_4)
plot.title("x_3 and x_4")
plot.xlabel('x_3')
plot.ylabel('x_4')
plot.show()

corr, _ = pearsonr(x_3, x_5)
print("Pearsons correlation of x_3 and x_5: %.3f" % corr)
plot.scatter(x_3, x_5)
plot.title("x_3 and x_5")
plot.xlabel('x_3')
plot.ylabel('x_5')
plot.show()

corr, _ = pearsonr(x_3, x_6)
print("Pearsons correlation of x_3 and x_6: %.3f" % corr)
plot.scatter(x_3, x_6)
plot.title("x_3 and x_6")
plot.xlabel('x_3')
plot.ylabel('x_6')
plot.show()

corr, _ = pearsonr(x_4, x_5)
print("Pearsons correlation of x_4 and x_5: %.3f" % corr)
plot.scatter(x_4, x_5)
plot.title("x_4 and x_5")
plot.xlabel('x_4')
plot.ylabel('x_5')
plot.show()

corr, _ = pearsonr(x_4, x_6)
print("Pearsons correlation of x_4 and x_6: %.3f" % corr)
plot.scatter(x_4, x_6)
plot.title("x_4 and x_6")
plot.xlabel('x_4')
plot.ylabel('x_6')
plot.show()

corr, _ = pearsonr(x_5, x_6)
print("Pearsons correlation of x_5 and x_6: %.3f" % corr)
plot.scatter(x_5, x_6)
plot.title("x_5 and x_6")
plot.xlabel('x_5')
plot.ylabel('x_6')
plot.show()
```
